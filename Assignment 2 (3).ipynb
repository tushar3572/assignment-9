```python
   #Answer: 3
    
Two notable exceptions derived from 'ArithmeticError' are 'ZeroDivisionError' and 'overflowError'

1. ZeroDivisionError: This exception is raised when an attempt is made to divide a number by zero.   
```

```python
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print("Exception caught:", e)
```

```python
2.OverflowError: This exception is raised when the result of an arithmetic operation exceeds the maximum representable value.
```

```python
import sys

try:
    result = sys.maxsize + 1
except OverflowError as e:
    print("Exception caught:", e)
```

```python
    #Answer: 4
    
LookupError Exception is the Base class for errors raised when something can’t be found.
The base class for the exceptions that are raised when a key or index used on a mapping or sequence is invalid:
IndexError, KeyError.

An IndexError is raised when a sequence reference is out of range.  
```

```python
dictionary = {'apple': 'red', 'banana': 'yellow', 'orange': 'orange'}

try:
    color = dictionary['grape']
except KeyError as e:
    print("Exception caught:", e)
```

```python
IndexError: This exception is raised when an index is out of range for a sequence (such as a list or string).
```

```python
my_list = [1, 2, 3]

try:
    value = my_list[5]
except IndexError as e:
    print("Exception caught:", e)
```
